```python
import math
class Shapes: #This is the base/parent class
    def __init__(self):
        pass #base constructor does nothing
    def area(self):#virtual method
        raise NotImplementedError #child class must implement this 
    
    def perimeter(self):
        raise NotImplementedError
    def perimeter_points(self):#virtual function to return up to 16 points
        raise NotImplementedError
        
        
class rectangle(Shapes):
    def __init__(self,length,width,x,y):
        #store length,width and bottom left corner coordinates
        self.__length=length
        self.__width=width
        self.__x=x
        self.__y=y
        #accessors:
    def get_length(self):
        return self.__length
    def get_width(self):
        return self.__width
    def get_x(self):
        return self.__x
    def get_y(self):
        return self.__y
    def area(self):
        return self.__length*self.__width
    def perimeter(self):
        return 2*(self.__length+self.__width)
    # Return up to 16 points along the perimeter
    def perimeter_points(self):
        points = [] # empty list to store points
        # divide each side into 4 equal segments:
        for i in range(4):
            #bottom side:
            points.append((self.__x + (i * self.__width / 4), self.__y))
            #right side:
            points.append((self.__x + self.__width, self.__y + (i * self.__length / 4)))
            #top side:
            points.append((self.__x + self.__width - (i * self.__width / 4), self.__y + self.__length))
            #left side:
            points.append((self.__x, self.__y + self.__length - (i * self.__length / 4)))
        return points #return all points
        
class circle(Shapes): #inheriting from Shapes
    def __init__(self,radius,x,y):
        #store radius and center coordinates
        self.__radius=radius
        self.__x=x
        self.__y=y
        #accessors:
    def get_radius(self):
        return self.__radius
    def get_x(self):
        return self.__x
    def get_y(self):
        return self.__y
    def area(self):
        #calculate area using pi * r^2
        return (math.pi)*(self.__radius*self.__radius)
    def perimeter(self):
        #calculate perimeter(circumference) using 2 * pi * r
        return 2*(math.pi)*self.__radius
    
    #Return up to 16 points along the circle perimeter:
    def perimeter_points(self):
        point=[]
        for i in range(0,16):
            angle=i*(2*math.pi/16) #loop 16 times to get evenly spaced points
            point_x=self.__x+self.__radius*math.cos(angle) #x=cx+r*cos(angle)
            point_y=self.__y+self.__radius*math.sin(angle) #y=cy+r*sin(angle)
            point.append((point_x,point_y)) #add point to list
        return point
    
    
class triangle(Shapes): #Triangle class inherits from Shapes
    def __init__(self,x1,y1,x2,y2,x3,y3):
        # store coordinates of three vertices:
        self.__x1=x1
        self.__y1=y1
        self.__x2=x2
        self.__y2=y2
        self.__x3=x3
        self.__y3=y3
    # accessors to read vertex coordinates:
    def get_firstcoordinate(self):
        return (self.__x1,self.__y1)
    def get_secondcoordinate(self):
        return (self.__x2,self.__y2)
    def get_thirdcoordinate(self):
        return (self.__x3,self.__y3)
    # calculating area using shoelace formula:
    def area(self):
        return abs(self.__x1*(self.__y2 - self.__y3) +self.__x2*(self.__y3 - self.__y1) +self.__x3*(self.__y1 - self.__y2)) / 2
    
    
    # calculate perimeter by summing side lengths:
    def perimeter(self):
        side1 = ((self.__x2-self.__x1)**2 + (self.__y2-self.__y1)**2)**0.5
        side2 = ((self.__x3-self.__x2)**2 + (self.__y3-self.__y2)**2)**0.5
        side3 = ((self.__x1-self.__x3)**2 + (self.__y1-self.__y3)**2)**0.5
        return side1 + side2 + side3
    
    #Return up to 16 points along the circle perimeter:
    def perimeter_points(self):
        points = []
        # defining edges as start and end points:
        edges = [((self.__x1, self.__y1), (self.__x2, self.__y2)),
            ((self.__x2, self.__y2), (self.__x3, self.__y3)),
            ((self.__x3, self.__y3), (self.__x1, self.__y1))]
        # distribute points along edges:
        points_per_edge = [5, 5, 6] ##distribute points along edges to total 16
        for i in range(3): #loop through edges
            start =edges[i][0]
            end =edges[i][1]
            n =points_per_edge[i]

            for t in range(n):
                f =t/(n - 1) #fraction along the edge
                x =start[0]+f*(end[0]-start[0])
                y =start[1]+f*(end[1]-start[1])
                points.append((x, y)) #add point

        return points
```

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
#CompoundShape class to hold multiple shapes together
class CompoundShape(Shapes):
    def __init__(self):
        self.shapes=[] #storing a list of shapes and their characters
        
    def add_shapes(self,shape,char='*'): #add a shape to the compound shape
        ## I am storing each shape as a tuple:(shape_object,charactertodraw)
        self.shapes.append((shape,char))
        
        
    def paint(self,canvas): #paint all shapes on a given canvas
        for shape, char in self.shapes: #loop over all shapes in the compound
            shape.paint(canvas, char) #call the paint method of each individual shape
    #this uses polymorphism: each shape knows how to paint itself
```

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height) #creating a canvas of the given width and height
        self.drawing_group = CompoundShape() #Creating a CompoundShape object to hold multiple shapes
    def add_shape(self, shape, char='*'): #adding a shape to the drawing
        self.drawing_group.add_shapes(shape, char) #I am using the compoundShape method to store the shape and its drawing character
    def render(self):
        self.canvas.clear_canvas() ##clearing the canvas before drawing
        self.drawing_group.paint(self.canvas) #painting all shapes from the compound group onto the canvas
        self.canvas.display() #now i am displaying
```

```python
class RasterDrawing:
    def __init__(self,width,height):
        self.canvas=Canvas(width,height) #creating a blank canvas
        self.shapes=[]  # here i am storing shapes here

    def add_shape(self,shape,char='*'):
        self.shapes.append((shape,char)) #adding a shape and its drawing character to the list

    def render(self):
        self.canvas.clear_canvas()  #here i am clearing
        for shape,char in self.shapes:
            #darwing each shape on the canvas
            shape.paint(self.canvas, char)
        self.canvas.display()  #here i am displaying
    def save(self, filename):
        #saving all shapes to a file so i can load later
        with open(filename, "w") as f:
            for shape, char in self.shapes:
                f.write(f"{repr(shape)},'{char}'\n") #using repr() of the shape and storing the character
```

```python
def raster_loader(filename, width=80, height=60):
    drawing = RasterDrawing(width, height) #creating empty drawing
    with open(filename,"r") as f:
        for line in f:
            #spliting the line into shape code and character
            shape_code,char=line.strip().rsplit(",", 1)
            shape = eval(shape_code) #recreating the shape object
            char = char.strip("'")   #removing quotes from the saved char
            drawing.add_shape(shape, char)
    return drawing
```
