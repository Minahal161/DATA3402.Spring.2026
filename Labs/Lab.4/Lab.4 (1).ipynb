```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        # Empty canvas is a matrix with element being the "space" character
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x,x+w):
            self.set_pixel(i,y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y,y+h):
            self.set_pixel(x,i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        slope = (y2-y1) / (x2-x1)
        for y in range(y1,y2):
            x= int(slope * y)
            self.set_pixel(x,y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))
```

```python
import math
class Shapes:
    def __init__(self):
        pass
    def area(self):
        raise NotImplementedError
    
    def perimeter(self):
        raise NotImplementedError
    def perimeter_points(self):
        raise NotImplementedError
    def inside(self,px,py):
        raise NotImplementedError
    def overlaps(self,other):
        for px,py in self.perimeter_points():
            if (other.inside(px,py)):
                return True
        for px,py in other.perimeter_points():
            if(self.inside(px,py)):
                return True
        return False
    def paint(self, canvas, char='*'):
        # Loop through every possible pixel on the canvas
        for r in range(canvas.height):
            for c in range(canvas.width):
                # If the current pixel (row, col) is inside the shape, set it!
                if self.inside(c, r): # x is column, y is row
                    canvas.set_pixel(r, c, char)
```

```python
class circle(Shapes): 
    def __init__(self,radius,x,y):
        self.__radius=radius
        self.__x=x
        self.__y=y
    def get_radius(self):
        return self.__radius
    def get_x(self):
        return self.__x
    def get_y(self):
        return self.__y
    def area(self):
        return (math.pi)*(self.__radius*self.__radius)
    def perimeter(self):
        return 2*(math.pi)*self.__radius
    
    def perimeter_points(self):
        point=[]
        for i in range(0,16):
            angle=i*(2*math.pi/16)
            point_x=self.__x+self.__radius*math.cos(angle)
            point_y=self.__y+self.__radius*math.sin(angle)
            point.append((point_x,point_y))
        return point
    def inside(self,px,py):
        if(((self.__x-px)**(2) + (self.__y-py)**(2) )**0.5 <=self.__radius):
            return True
        else:
            return False
```

```python
class triangle(Shapes):
    def __init__(self,x1,y1,x2,y2,x3,y3):
        self.__x1=x1
        self.__y1=y1
        self.__x2=x2
        self.__y2=y2
        self.__x3=x3
        self.__y3=y3
    def get_firstcoordinate(self):
        return (self.__x1,self.__y1)
    def get_secondcoordinate(self):
        return (self.__x2,self.__y2)
    def get_thirdcoordinate(self):
        return (self.__x3,self.__y3)
    def area(self):
        return abs(self.__x1*(self.__y2 - self.__y3) +self.__x2*(self.__y3 - self.__y1) +self.__x3*(self.__y1 - self.__y2)) / 2
    def perimeter(self):
        side1 = ((self.__x2-self.__x1)**2+(self.__y2-self.__y1)**2)**0.5
        side2 = ((self.__x3-self.__x2)**2+(self.__y3-self.__y2)**2)**0.5
        side3 = ((self.__x1-self.__x3)**2+(self.__y1-self.__y3)**2)**0.5
        return side1 + side2 + side3
    def perimeter_points(self):
        points = []
        edges = [((self.__x1, self.__y1), (self.__x2, self.__y2)),
            ((self.__x2, self.__y2), (self.__x3, self.__y3)),
            ((self.__x3, self.__y3), (self.__x1, self.__y1))]
        points_per_edge = [5, 5, 6]  
        for i in range(3):
                start = edges[i][0]
                end = edges[i][1]
                n = points_per_edge[i]

                for t in range(n):
                    f = t / (n - 1)
                    x = start[0] + f * (end[0] - start[0])
                    y = start[1] + f * (end[1] - start[1])
                    points.append((x, y))

        return points
    def inside(self, px, py):
        def tri_area(x1, y1, x2, y2, x3, y3):
            return abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2)) / 2
        main_area = tri_area(self.__x1,self.__y1,self.__x2,self.__y2,self.__x3,self.__y3)
        area1 = tri_area(px,py,self.__x2,self.__y2,self.__x3,self.__y3)
        area2 = tri_area(self.__x1,self.__y1,px, py,self.__x3,self.__y3)
        area3 = tri_area(self.__x1,self.__y1,self.__x2,self.__y2,px,py)
        if abs((area1 + area2 + area3)-main_area) < 1e-6:
            return True
        else:
            return False
        
```

```python
class CompoundShape(Shapes):
    def __init__(self):
        self.shapes = []
    def add_shapes(self, shape, char='*'):
        self.shapes.append((shape, char))
    def paint(self, canvas):
        for shape, char in self.shapes:
            shape.paint(canvas, char)
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.drawing_group = CompoundShape()
    def add_shape(self, shape, char='*'):
        self.drawing_group.add_shapes(shape, char)
    def render(self):
        self.canvas.clear_canvas()
        self.drawing_group.paint(self.canvas)
        self.canvas.display()
```

```python
my_art = RasterDrawing(60, 20)
rect1 = rectangle(5, 10, 5, 5)
circ1 = circle(4, 40, 10)
my_art.add_shape(rect1, char='#')
my_art.add_shape(circ1, char='@')
print("First Render")
my_art.render()
tri1 = triangle(15, 18, 25, 18, 20, 12)
my_art.add_shape(tri1, char='^')
print(" Second Render(Modified with a Triangle)")
my_art.render()
```

```python
class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []  # <-- store shapes here

    def add_shape(self, shape, char='*'):
        self.shapes.append((shape, char))

    def render(self):
        self.canvas.clear_canvas()
        for shape, char in self.shapes:
            shape.paint(self.canvas, char)
        self.canvas.display()
    def save(self, filename):
        with open(filename, "w") as f:
            for shape, char in self.shapes:
                f.write(f"{repr(shape)},'{char}'\n")
```

```python
def raster_loader(filename, width=80, height=60):
    drawing = RasterDrawing(width, height)
    with open(filename, "r") as f:
        for line in f:
            shape_code, char = line.strip().rsplit(",", 1)
            shape = eval(shape_code)
            char = char.strip("'")
            drawing.add_shape(shape, char)
    return drawing
```

```python
drawing = RasterDrawing(80, 60)
drawing.add_shape(rectangle(10, 5, 2, 2), '#')
drawing.add_shape(circle(4, 20, 10), '@')
drawing.render()  
drawing.save("my_drawing.txt")

loaded_drawing = raster_loader("my_drawing.txt", 80, 60)
loaded_drawing.render()  
```
