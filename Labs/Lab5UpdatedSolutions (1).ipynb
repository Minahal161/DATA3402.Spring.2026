1. Create a `matrix` class with the following properties:
    * It can be initialized in 2 ways:
        1. with arguments `n` and `m`, the size of the matrix. A newly instanciated matrix will contain all zeros.
        2. with a list of lists of values. Note that since we are using lists of lists to implement matrices, it is possible that not all rows have the same number of columns. Test explicitly that the matrix is properly specified.
    * Matrix instances `M` can be indexed with `M[i][j]` and `M[i,j]`.
    * Matrix assignment works in 2 ways:
        1. If `M_1` and `M_2` are `matrix` instances `M_1=M_2` sets the values of `M_1` to those of `M_2`, if they are the same size. Error otherwise.
        2. In example above `M_2` can be a list of lists of correct size.


```python
class matrix:
    def __init__(self,n,m=None): #constructor for creating a matrix
        #these are the dimensions of the matrix which are temporary until we determine how it was initialised
        self.n=n
        self.m=m
        if isinstance(m,int):
            self.matrix = [] #list that will store all rows
            for i in range(self.n): #creating n rows
                row=[]
                for j in range(self.m): #here row contains m columns initialized to 0
                    row.append(0)
                self.matrix.append(row) #add row to matrix
        if isinstance(n,list) and m==None:
            if len(n)==0: #checking matrix is not empty
                raise ValueError("Matrix cannot be empty")
            row_len=len(n[0]) #here i am getting number of columns from first row
            for row in n:
                if len(row) != row_len: #checking that all rows have same number of columns
                    print("All rows must have the same length")
            # storing dimensions:
            self.n = len(n)
            self.m = row_len
            #here i am copying the data into the matrix
            self.matrix = []         
            for row in n:
                self.matrix.append(list(row))
    
    def __getitem__(self,key):
        if isinstance(key,tuple): #for M[i,j]
            return self.matrix[key[0]][key[1]]
        return self.matrix[key] #for M[i]
    
    
    def __setitem__(self, key, value): # this is allowing setting values like M[i,j]=value
        if isinstance(key,tuple):   #if index is a tuple which is M[i,j]
            r, c = key
            self.matrix[r][c] = value
            #if assigning a whole row:
        else:
            self.matrix[key] = value
            
            
            
    def assign(self, other): #This allows assigning values from another matrix or list of lists
        #if assigning from another matrix object:
        if isinstance(other, matrix):
            data = other.matrix
        #if assigning from list of lists:
        elif isinstance(other, list):
            data = other
        else:
            raise ValueError("Assignment can only be done from a matrix or list of lists")
        if len(data) != self.n or len(data[0]) != self.m:   #checking size
            raise ValueError("Assignment cannot be done due to size mismatch")
            
        self.matrix = []         ##copying the values into the matrix
        for row in data: 
            self.matrix.append(list(row))
    def display(self):      #here i am displayingg matrix
        for row in self.matrix:
            print(row)     
    
```

```python
class matrix:
    def __init__(self,n,m=None): #constructor for creating a matrix
        #these are the dimensions of the matrix which are temporary until we determine how it was initialised
        self.n=n
        self.m=m
        if isinstance(m,int):
            self.matrix = [] #list that will store all rows
            for i in range(self.n): #creating n rows
                row=[]
                for j in range(self.m): #here row contains m columns initialized to 0
                    row.append(0)
                self.matrix.append(row) #add row to matrix
        if isinstance(n,list) and m==None:
            if len(n)==0: #checking matrix is not empty
                raise ValueError("Matrix cannot be empty")
            row_len=len(n[0]) #here i am getting number of columns from first row
            for row in n:
                if len(row) != row_len: #checking that all rows have same number of columns
                    print("All rows must have the same length")
            # storing dimensions:
            self.n = len(n)
            self.m = row_len
            #here i am copying the data into the matrix
            self.matrix = []         
            for row in n:
                self.matrix.append(list(row))
    
    def __getitem__(self,key):
        if isinstance(key, tuple):  #for M[i,j]
            row_key,col_key=key
            
            #for slicing M[a:b , c:d]:
        if isinstance(row_key, slice) and isinstance(col_key, slice):
             new_rows = []  #creating a list to store the rows of the new submatrix
             for i in range(row_key.start, row_key.stop):
                new_rows.append(self.matrix[i][col_key.start:col_key.stop]) #extracting the selected columns from each row

             return matrix(new_rows)
         #if both row and column are integers:
        elif isinstance(row_key, int) and isinstance(col_key, int):
            return self.matrix[row_key][col_key]    #return the element at position (row_key,col_key)
        
        
          #if key is a single integer:
        elif isinstance(key, int):  #for M[i]
            return self.matrix[key]
        
        
    def __setitem__(self,key,value): # this is allowing setting values like M[i,j]=value
        if isinstance(key,tuple):   #if index is a tuple which is M[i,j]
            r,c =key
            self.matrix[r][c]=value
            #if assigning a whole row:
        else:
            self.matrix[key] = value
    def assign(self, other): #This allows assigning values from another matrix or list of lists
        #if assigning from another matrix object:
        if isinstance(other, matrix):
            data = other.matrix
        #if assigning from list of lists:
        elif isinstance(other, list):
            data = other
        else:
            raise ValueError("Assignment can only be done from a matrix or list of lists")
        if len(data) != self.n or len(data[0]) != self.m:   #checking size
            raise ValueError("Assignment cannot be done due to size mismatch")
            
        self.matrix = []         ##copying the values into the matrix
        for row in data: 
            self.matrix.append(list(row)) #appending to the list
            
            
    def display(self):      #here i am displayingg matrix
        for row in self.matrix:
            print(row)     
    
    def shape(self):  #returns matrix shape as tuple
            return self.n,self.m
        
    def transpose(self):  #returns transpose of matrix
        newmatrix=matrix(self.m,self.n)  #creating a new matrix
        for i in range(self.n):     #looping through each element in the original matrix
            for j in range(self.m):
                newmatrix.matrix[j][i]=self.matrix[i][j]  #here i am swapping row and column indices
        return newmatrix   #returning transposed matrix
    
    
    def row(self,n):  #returns nth row
        return matrix([self.matrix[n]])
    def column(self,n):    #returns nth column
        return matrix([[row[n]] for row in self.matrix])
    
    def to_list(self):  #matrix is already stored as list of lists
        return self.matrix    #returns list
    
    def block(self,n0,n1,m0,m1):
        subrows=[]
        for i in range(m0,m1):  #looping through the selected rows
            subrows.append(self.matrix[i][n0:n1])   ##slicing the selected columns from each row
        return matrix(subrows)   #creating and return a new matrix from the extracted rows
    
```

3. Write functions that create special matrices (note these are standalone functions, not member functions of your `matrix` class):
    * `constant(n,m,c)`: returns a `n` by `m` matrix filled with floats of value `c`.
    * `zeros(n,m)` and `ones(n,m)`: return `n` by `m` matrices filled with floats of value `0` and `1`, respectively.
    * `eye(n)`: returns the n by n identity matrix.

```python
def constant(n,m,c):
    constantmatrix=[]
    for i in range(n):   #looping through rows
        row=[]
        for j in range(m):  #looping through columns
            row.append(float(c))  #hee converting to float and then appending to list
        constantmatrix.append(row)
        
    return matrix(constantmatrix)  #converting list of lists into a matrix object


def zeros(n,m):
    zerosmatrix=[]
    for i in range(n): #looping through rows
        row=[]
        for j in range(m): #looping through columns
            row.append(0)    #appending 0s to the row lists
        zerosmatrix.append(row)   #here i am adding row to matrix
        
    return matrix(zerosmatrix)   #returning matrix object


def ones(n,m):
    onesmatrix=[]
    for i in range(n):   #looping through rows
        row=[]
        for j in range(m):   #looping through columns
            row.append(1)   #appending 1s to the row lists
        onesmatrix.append(row)  #here i am adding row to matrix
        
    return matrix(onesmatrix) #returning matrix object
def eye(n):
    #here i am creating an n by n identity matrix
    identitymatrix=[]
    for i in range(n): 
        row=[]
        for j in range(n):
            #diagonal elements=1
            #all other elements=0
            if(i==j):
                row.append(1)
            else:
                row.append(0)  #adding row to matrix
        identitymatrix.append(row)
    return matrix(identitymatrix)   #returning the identity matrix
                
```

```python
class matrix:
    def __init__(self,n,m=None): #constructor for creating a matrix
        #these are the dimensions of the matrix which are temporary until we determine how it was initialised
        self.n=n
        self.m=m
        if isinstance(m,int):
            self.matrix = [] #list that will store all rows
            for i in range(self.n): #creating n rows
                row=[]
                for j in range(self.m): #here row contains m columns initialized to 0
                    row.append(0)
                self.matrix.append(row) #add row to matrix
        if isinstance(n,list) and m==None:
            if len(n)==0: #checking matrix is not empty
                raise ValueError("Matrix cannot be empty")
            row_len=len(n[0]) #here i am getting number of columns from first row
            for row in n:
                if len(row) != row_len: #checking that all rows have same number of columns
                    print("All rows must have the same length")
            # storing dimensions:
            self.n = len(n)
            self.m = row_len
            #here i am copying the data into the matrix
            self.matrix = []         
            for row in n:
                self.matrix.append(list(row))
    
    def __getitem__(self,key):
        if isinstance(key, tuple):  #for M[i,j]
            row_key,col_key=key
            
            #for slicing M[a:b , c:d]:
        if isinstance(row_key, slice) and isinstance(col_key, slice):
             new_rows = []  #creating a list to store the rows of the new submatrix
             for i in range(row_key.start, row_key.stop):
                new_rows.append(self.matrix[i][col_key.start:col_key.stop]) #extracting the selected columns from each row

             return matrix(new_rows)
         #if both row and column are integers:
        elif isinstance(row_key, int) and isinstance(col_key, int):
            return self.matrix[row_key][col_key]    #return the element at position (row_key,col_key)
        
        
          #if key is a single integer:
        elif isinstance(key,int):  #for M[i]
            return self.matrix[key]
        
        
    def __setitem__(self,key,value): # this is allowing setting values like M[i,j]=value
        if isinstance(key,tuple):   #if index is a tuple which is M[i,j]
            r,c =key
            self.matrix[r][c]=value
            #if assigning a whole row:
        else:
            self.matrix[key] = value
    def assign(self, other): #This allows assigning values from another matrix or list of lists
        #if assigning from another matrix object:
        if isinstance(other, matrix):
            data = other.matrix
        #if assigning from list of lists:
        elif isinstance(other, list):
            data = other
        else:
            raise ValueError("Assignment can only be done from a matrix or list of lists")
        if len(data) != self.n or len(data[0]) != self.m:   #checking size
            raise ValueError("Assignment cannot be done due to size mismatch")
            
        self.matrix = []         ##copying the values into the matrix
        for row in data: 
            self.matrix.append(list(row)) #appending to the list
            
            
    def display(self):      #here i am displayingg matrix
        for row in self.matrix:
            print(row)     
    
    def shape(self):  #returns matrix shape as tuple
            return self.n,self.m
        
    def transpose(self):  #returns transpose of matrix
        newmatrix=matrix(self.m,self.n)  #creating a new matrix
        for i in range(self.n):     #looping through each element in the original matrix
            for j in range(self.m):
                newmatrix.matrix[j][i]=self.matrix[i][j]  #here i am swapping row and column indices
        return newmatrix   #returning transposed matrix
    
    
    def row(self,n):  #returns nth row
        return matrix([self.matrix[n]])
    def column(self,n):    #returns nth column
        return matrix([[row[n]] for row in self.matrix])
    
    def to_list(self):  #matrix is already stored as list of lists
        return self.matrix    #returns list
    
    def block(self,n0,n1,m0,m1):
        subrows=[]
        for i in range(m0,m1):  #looping through the selected rows
            subrows.append(self.matrix[i][n0:n1])   ##slicing the selected columns from each row
        return matrix(subrows)   #creating and return a new matrix from the extracted rows
    
    def scalarmul(self,c):
          #multiplying every element of the matrix by scalar c
        scalarmul=matrix(self.n,self.m)
        for i in range(self.n):  #looping through rows
            for j in range(self.m):
                scalarmul.matrix[i][j]=self.matrix[i][j]*c  #here i am multipliying
        return scalarmul
    
    def add(self,N):    #elementwise addition of two matrices
        if(self.n!=N.n or self.m!=N.m):
            raise ValueError("Matrices must have same dimensions")
        else:
            addmatrix=[]
            for i in range(self.n):
                row=[]
                for j in range(self.m):
                    final=self.matrix[i][j]+N.matrix[i][j]
                    row.append(final)
                addmatrix.append(row)
            return matrix(addmatrix)   
        
    def sub(self,N):  #element wise subtraction of two matrices
        if(self.n!=N.n or self.m!=N.m):
            raise ValueError("Matrices must have same dimensions")
        else:
            submatrix=[]
            for i in range(self.n):
                row=[]
                for j in range(self.m):
                    final=self.matrix[i][j]-N.matrix[i][j] #here i am subtracting elements
                    row.append(final)
                submatrix.append(row)
            return matrix(submatrix)
                            
           
                        
    def element_mult(self,N):  #element wise multiplication of two matrices
        if(self.n!=N.n or self.m!=N.m):
            raise ValueError("Matrices must have same dimensions")
        else:
            elementmult=[]
            for i in range(self.n):
                row=[]
                for j in range(self.m):
                    final=self.matrix[i][j]*N.matrix[i][j]
                    row.append(final)
                elementmult.append(row)
            return matrix(elementmult)
        
    def mat_mult(self,N):  #matrix multiplication
        if self.m!=N.n:
            raise ValueError("Matrix dimensions not correct for multiplication")
        matmult=[]
        for i in range(self.n):
            row =[]
            for j in range(N.m):
                total =0
                for k in range(self.m):
                    total+=self.matrix[i][k]*N.matrix[k][j]
                row.append(total)
            matmult.append(row)
    
        return matrix(matmult)
         
    def equals(self,N):  #checking if two matrices are equal
        if(self.n!=N.n or self.m!=N.m):   #here i am checking dimensions if they do not match then False
            return False
        else:
            for i in range(self.n):
                for j in range(self.m):
                    if(self.matrix[i][j]!=N.matrix[i][j]):
                        return False
            return True
    
```

```python
class matrix:
    def __init__(self,n,m=None): #constructor for creating a matrix
        #these are the dimensions of the matrix which are temporary until we determine how it was initialised
        self.n=n
        self.m=m
        if isinstance(m,int):
            self.matrix = [] #list that will store all rows
            for i in range(self.n): #creating n rows
                row=[]
                for j in range(self.m): #here row contains m columns initialized to 0
                    row.append(0)
                self.matrix.append(row) #add row to matrix
        if isinstance(n,list) and m==None:
            if len(n)==0: #checking matrix is not empty
                raise ValueError("Matrix cannot be empty")
            row_len=len(n[0]) #here i am getting number of columns from first row
            for row in n:
                if len(row) != row_len: #checking that all rows have same number of columns
                    print("All rows must have the same length")
            # storing dimensions:
            self.n = len(n)
            self.m = row_len
            #here i am copying the data into the matrix
            self.matrix = []         
            for row in n:
                self.matrix.append(list(row))
    
    def __getitem__(self,key):
        if isinstance(key, tuple):  #for M[i,j]
            row_key,col_key=key
            
            #for slicing M[a:b , c:d]:
        if isinstance(row_key, slice) and isinstance(col_key, slice):
             new_rows = []  #creating a list to store the rows of the new submatrix
             for i in range(row_key.start, row_key.stop):
                new_rows.append(self.matrix[i][col_key.start:col_key.stop]) #extracting the selected columns from each row

             return matrix(new_rows)
         #if both row and column are integers:
        elif isinstance(row_key, int) and isinstance(col_key, int):
            return self.matrix[row_key][col_key]    #return the element at position (row_key,col_key)
        
        
          #if key is a single integer:
        elif isinstance(key,int):  #for M[i]
            return self.matrix[key]
        
        
    def __setitem__(self,key,value): # this is allowing setting values like M[i,j]=value
        if isinstance(key,tuple):   #if index is a tuple which is M[i,j]
            r,c =key
            self.matrix[r][c]=value
            #if assigning a whole row:
        else:
            self.matrix[key] = value
    def assign(self, other): #This allows assigning values from another matrix or list of lists
        #if assigning from another matrix object:
        if isinstance(other, matrix):
            data = other.matrix
        #if assigning from list of lists:
        elif isinstance(other, list):
            data = other
        else:
            raise ValueError("Assignment can only be done from a matrix or list of lists")
        if len(data) != self.n or len(data[0]) != self.m:   #checking size
            raise ValueError("Assignment cannot be done due to size mismatch")
            
        self.matrix = []         ##copying the values into the matrix
        for row in data: 
            self.matrix.append(list(row)) #appending to the list
            
            
    def display(self):      #here i am displayingg matrix
        for row in self.matrix:
            print(row)     
    
    def shape(self):  #returns matrix shape as tuple
            return self.n,self.m
        
    def transpose(self):  #returns transpose of matrix
        newmatrix=matrix(self.m,self.n)  #creating a new matrix
        for i in range(self.n):     #looping through each element in the original matrix
            for j in range(self.m):
                newmatrix.matrix[j][i]=self.matrix[i][j]  #here i am swapping row and column indices
        return newmatrix   #returning transposed matrix
    
    
    def row(self,n):  #returns nth row
        return matrix([self.matrix[n]])
    def column(self,n):    #returns nth column
        return matrix([[row[n]] for row in self.matrix])
    
    def to_list(self):  #matrix is already stored as list of lists
        return self.matrix    #returns list
    
    def block(self,n0,n1,m0,m1):
        subrows=[]
        for i in range(m0,m1):  #looping through the selected rows
            subrows.append(self.matrix[i][n0:n1])   ##slicing the selected columns from each row
        return matrix(subrows)   #creating and return a new matrix from the extracted rows
    
    def scalarmul(self,c):
          #multiplying every element of the matrix by scalar c
        scalarmul=matrix(self.n,self.m)
        for i in range(self.n):  #looping through rows
            for j in range(self.m):
                scalarmul.matrix[i][j]=self.matrix[i][j]*c  #here i am multipliying
        return scalarmul
    
    def add(self,N):    #elementwise addition of two matrices
        if(self.n!=N.n or self.m!=N.m):
            raise ValueError("Matrices must have same dimensions")
        else:
            addmatrix=[]
            for i in range(self.n):
                row=[]
                for j in range(self.m):
                    final=self.matrix[i][j]+N.matrix[i][j]
                    row.append(final)
                addmatrix.append(row)
            return matrix(addmatrix)   
        
    def sub(self,N):  #element wise subtraction of two matrices
        if(self.n!=N.n or self.m!=N.m):
            raise ValueError("Matrices must have same dimensions")
        else:
            submatrix=[]
            for i in range(self.n):
                row=[]
                for j in range(self.m):
                    final=self.matrix[i][j]-N.matrix[i][j] #here i am subtracting elements
                    row.append(final)
                submatrix.append(row)
            return matrix(submatrix)
                            
           
                        
    def element_mult(self,N):  #element wise multiplication of two matrices
        if(self.n!=N.n or self.m!=N.m):
            raise ValueError("Matrices must have same dimensions")
        else:
            elementmult=[]
            for i in range(self.n):
                row=[]
                for j in range(self.m):
                    final=self.matrix[i][j]*N.matrix[i][j]
                    row.append(final)
                elementmult.append(row)
            return matrix(elementmult)
        
    def mat_mult(self,N):  #matrix multiplication
        if self.m!=N.n:
            raise ValueError("Matrix dimensions not correct for multiplication")
        matmult=[]
        for i in range(self.n):
            row =[]
            for j in range(N.m):
                total =0
                for k in range(self.m):
                    total+=self.matrix[i][k]*N.matrix[k][j]
                row.append(total)
            matmult.append(row)
    
        return matrix(matmult)
         
    def equals(self,N):  #checking if two matrices are equal
        if(self.n!=N.n or self.m!=N.m):   #here i am checking dimensions if they do not match then False
            return False
        else:
            for i in range(self.n):
                for j in range(self.m):
                    if(self.matrix[i][j]!=N.matrix[i][j]):
                        return False
            return True
    def display(self):  
        for row in self.matrix:
            print(row) 
    
       
    def __add__(self,other):  #for M+N
        return self.add(other)  #will call the add method and will returns a new matrix with elementwise sum

    def __sub__(self,other):  #for M-N
        return self.sub(other) #will call the sub method and will returns a new matrix with elementwise subtraction

    
    def __scalarm__(self,other):  #for M*scalar
        return self.scalarmul(other)  #will return by multiplying every element by other 
    
    def __eq__(self,other):  #for M==N
        return self.equals(other) ##will return True if dimensions and elements match False otherwise
    
    def __mul__(self,other):
        if isinstance(other,(int,float)):   
            return self.scalarmul(other) #scalar multiplication if other is a number
        elif isinstance(other, matrix):       
            return self.mat_mult(other) #matrix multiplication if other is a matrix
        else:
            raise ValueError("Type is not correct for multiplication")
    def __rmul__(self, other):
        return self.__mul__(other)     #just calls __mul__ allowing scalar*matrix
    
```
