````python
# Write yourrr solution here
def full_tictactoe(n):
    board = [[0 for _ in range(n)] for _ in range(n)]
    player = 1
    
    while True:
        print(f"\nPlayer {player}'s turn:")
        location_boardfinal(board)
        valid(board, player)
        status = checkerfinal(board)
        
        if status == 1:
            print("Player 1 (X) wins!")
            location_boardfinal(board)
            break
        elif status == 2:
            print("Player 2 (O) wins!")
            location_boardfinal(board)
            break
        else:
            empty_found = False
            for row in board:
                if 0 in row:
                    empty_found = True

            if not empty_found:
                print("Game is a draw!")
                location_boardfinal(board)
                break

        if player == 1:
            player=2
        else:
            player=1
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
def evaluate(board):
    result = checkerfinal(board)
    if result == 2:
        return 1
    elif result == 1:
        return -1
    else:
        return 0
def minimax(board, depth, is_maximizing):
    score = evaluate(board)
    if score != 0 or depth == 0 or not any(0 in row for row in board):
        return score

    n = len(board)
    if is_maximizing:
        best = -float('inf')
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 2
                    best = max(best, minimax(board, depth-1, False))
                    board[i][j] = 0
        return best
    else:  
        best = float('inf')
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 1
                    best = min(best, minimax(board, depth-1, True))
                    board[i][j] = 0
        return best


def computer_move_depth(board, max_depth=3):
    n = len(board)
    best_score = -float('inf')
    best_move = None

    for i in range(n):
        for j in range(n):
            if board[i][j] == 0:
                board[i][j] = 2  
                score = minimax(board, max_depth-1, False)
                board[i][j] = 0
                if score > best_score:
                    best_score = score
                    best_move = (i, j)

    if best_move:
        i, j = best_move
        board[i][j] = 2
        print(f"Computer moves to {chr(65+i)}{j+1}")
def tic_tac_toe_vs_computer_depth(n, max_depth=3):
    
    board = [[0 for _ in range(n)] for _ in range(n)]
    player = 1 

    while True:
        print("\nCurrent Board:")
        location_boardfinal(board)

        if player == 1:
            valid(board, 1)
        else:
            print("Computer is thinking...")
            computer_move_depth(board, max_depth)
        status = checkerfinal(board)
        if status == 1:
            location_boardfinal(board)
            print("Player 1 (X) wins!")
            break
        elif status == 2:
            location_boardfinal(board)
            print("Computer (O) wins!")
            break
        elif not any(0 in row for row in board):
            location_boardfinal(board)
            print("Game is a draw!")
            break
        if player == 1:
            player=2
        else:
            player=1

````

````python
import random
def computer_move_maxdepth(board, player_to_move, max_depth):
    n = len(board)
    def minimax(b, depth, is_maximizing):
        status = checkerfinal(b)
        if status == 2: return 10 + depth   
        if status == 1: return -10 - depth 
        if not any(0 in row for row in b) or depth == 0:
            return 0
        
        if is_maximizing:
            best = -float('inf')
            for i in range(n):
                for j in range(n):
                    if b[i][j] == 0:
                        b[i][j] = 2
                        score = minimax(b, depth - 1, False)
                        b[i][j] = 0
                        best = max(best, score)
            return best
        else:
            best = float('inf')
            for i in range(n):
                for j in range(n):
                    if b[i][j] == 0:
                        b[i][j] = 1
                        score = minimax(b, depth - 1, True)
                        b[i][j] = 0
                        best = min(best, score)
            return best
    best_move = None
    if player_to_move == 2:
        best_score = -float('inf')
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 2
                    score = minimax(board, max_depth - 1, False)
                    board[i][j] = 0
                    if score > best_score:
                        best_score = score
                        best_move = (i, j)
    else:
        best_score = float('inf')
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 1
                    score = minimax(board, max_depth - 1, True)
                    board[i][j] = 0
                    if score < best_score:
                        best_score = score
                        best_move = (i, j)
    
    if best_move:
        board[best_move[0]][best_move[1]] = player_to_move
    else:
        empty = [(i,j) for i in range(n) for j in range(n) if board[i][j]==0]
        if empty:
            r, c = random.choice(empty)
            board[r][c] = player_to_move


def run_experiment():
    for size in [3, 4, 5]:
        wins1, wins2, draws = 0, 0, 0
        d1 = 3 if size < 5 else 2
        d2 = 1 

        for _ in range(10):
            board = [[0 for _ in range(size)] for _ in range(size)]
            turn = 1
            while True:
                current_depth = d1 if turn == 1 else d2
                computer_move_maxdepth(board, turn, current_depth)
                
                status = checkerfinal(board)
                if status == 1:
                    wins1 += 1
                    break
                if status == 2:
                    wins2 += 1
                    break
                if not any(0 in row for row in board):
                    draws += 1
                    break
                turn = 2 if turn == 1 else 1
        
        print(f"Results for {size}x{size}:")
        print(f"  Player 1 (Smarter, Depth {d1}) Wins: {wins1}")
        print(f"  Player 2 (Weaker, Depth {d2}) Wins: {wins2}")
        print(f"  Draws: {draws}")
        print(f"  Smarter Win Rate: {(wins1/10)*100}%\n")

if __name__ == "__main__":
    run_experiment()
````
